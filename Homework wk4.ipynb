### EDA of world record mile dataset

```python
from ggplot import *
import pandas as pd
DATA_DIR = '../data/'

%matplotlib inline
```

```python
wrmile = pd.read_csv(DATA_DIR + "world-record-mile.csv")
wrmile.head()
```

```python
ggplot(wrmile, aes(x = 'Year', y = 'seconds')) + \
    geom_point() + \
    ggtitle("(Year, Seconds) Scatterplot")
```

There are clearly two different types of data being represented here...

```python
ggplot(wrmile, aes(x = 'Year', y = 'seconds', color = 'Gender')) + \
    geom_point() + \
    ggtitle("(Year, Seconds) Heatmap")
```

We can see that there is a clear distinction between male and female athletes, with male athletes running a faster world record mile for a given year compared to females. 

We can also see that over time both men and women have been running faster miles.

```python
wrmile.info()
```

Above count shows that there is no missing data

```python
wrmile.describe()
```

```python
ggplot(wrmile, aes(x='Year')) + geom_histogram(binwidth=5, 
                                               colour="black", fill="green")
```

There is one time period in the early 1980s where more mile world records were set than in any other time period

```python
ggplot(wrmile, aes(x='seconds')) + geom_density()
```

```python
# Add a line for the mean:
avg_time = np.mean(wrmile.seconds)

ggplot(wrmile, aes(x='seconds')) + \
    geom_histogram(aes(y='..density..'),      # Histogram with density instead of count on y-axis
                   binwidth=5,
                   colour="black",
                   fill="green") + \
    geom_vline(aes(xintercept=avg_time),
               color="red",
               linetype="dashed",
               size=2)
```

```python
ggplot(wrmile, aes(x='seconds', fill='Gender')) + \
    geom_histogram(binwidth=5, alpha=.5, position="identity")
```

This histogram shows again that over the last century men have had faster mile times than women

```python
ggplot(wrmile, aes(x='seconds', colour='Gender')) + geom_density()
```

The above density plot shows that the world record time for men has improved only marginally (although with a high frequency), while the world record time for women has improved more significantly in terms of time reduction.
